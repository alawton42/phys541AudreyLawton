#### Question 2
Modify graph_runge_kutta_example.ipynb (in the Week5 folder) to solve the following differential equation initial value problem:
dy/dt = t y2
y(0) = 1
You can use WolframAlpha, for example, to get the exact solution.  Evaluate (if possible) the following:
a)	y(1)
b)	y(1.4)
c)	y(2)


```python
import numpy as np
import matplotlib.pyplot as plt
 
def f(y,t):
    return t*y**2

def y(t):
    return -2/(t**2-2.0)
```

```python
t0 = 0.0
y0 = y(t0)

print(y0)

tau = 1
```

```python
# First Order RK

y1 = y0 + tau*f(y0,t0)
print(f"First Order RK: y1_est = %.5f, y1_exact = %.5f, error = %.3f" % (y1,y(t0+tau), np.abs(y1-y(t0+tau))*100/y(t0+tau)))
```

```python
# Second Order RK

k1 = f(y0,t0)
k2 = f(y0+tau*k1,t0+tau)

y1 = y0 + tau*(k1+k2)/2.0
print(f"Second Order RK: y1_est = %.5f, y1_exact = %.5f, error = %.3f" % (y1,y(t0+tau), np.abs(y1-y(t0+tau))*100/y(t0+tau)))
```

```python
# Third Order RK

k1 = f(y0,t0)
k2 = f(y0+tau*k1/2.0,t0+tau/2.0)
k3 = f(y0+tau*k2,t0+tau)

y1 = y0 + tau*(k1+4*k2+k3)/6.0
print(f"Third Order RK: y1_est = %.5f, y1_exact = %.5f, error = %.3f" % (y1,y(t0+tau), np.abs(y1-y(t0+tau))*100/y(t0+tau)))
```

```python
# Fourth Order RK

k1 = f(y0,t0)
k2 = f(y0+tau*k1/2.0,t0+tau/2.0)
k3 = f(y0+tau*k2/2.0,t0+tau/2.0)
k4 = f(y0+tau*k3,t0+tau)

print(k1,k2,k3,k4)

y1 = y0 + tau*(k1+2*k2+2*k3+k4)/6.0

print(f"Fourth Order RK: y1_est = %.5f, y1_exact = %.5f, error = %.3f" % (y1,y(t0+tau), np.abs(y1-y(t0+tau))*100/y(t0+tau)))
```

```python
time = np.linspace(t0,t0+tau,100)
y_exact = y(time)

yk1 = y0 + time*k1
yk2 = y0 + time*k2
yk3 = y0 + time*k3
yk4 = y0 + time*k4

yk1pt = y0 + tau*k1
yk2pt = y0 + tau*k2
yk3pt = y0 + tau*k3
yk4pt = y0 + tau*k4
tau_array = np.array([t0+tau,t0+tau,t0+tau,t0+tau])
ypt_array = np.array([yk1pt,yk2pt,yk3pt,yk4pt])

tau_exact_pt = np.array([t0+tau])
y_exact_pt = np.array([y(t0+tau)])

tau_est_pt = np.array([t0+tau])
y_est_pt = np.array([y1])


plt.plot(time,y_exact,'k-',label='Exact')
plt.plot(time,yk1,'r--',label='y0 + t*k1')
plt.plot(time,yk2,'g--',label='y0 + t*k2')
plt.plot(time,yk3,'b--',label='y0 + t*k3')
plt.plot(time,yk4,'m--',label='y0 + t*k4')
plt.plot(tau_array,ypt_array,'ko',label='y0 + (t0+tau)*k Estimation')
plt.plot(tau_exact_pt,y_exact_pt,'ro',label='y1 exact')
plt.plot(tau_est_pt,y_est_pt,'bo',label='y1 est')
plt.ylim(0.8)

plt.legend()
plt.xlabel('t')
plt.ylabel('y')
```

```python
tau_b = 1.4
# Fourth Order RK for B) y(1.4)

k1 = f(y0,t0)
k2 = f(y0+tau_b*k1/2.0,t0+tau_b/2.0)
k3 = f(y0+tau_b*k2/2.0,t0+tau_b/2.0)
k4 = f(y0+tau_b*k3,t0+tau_b)

print(k1,k2,k3,k4)

y1 = y0 + tau_b*(k1+2*k2+2*k3+k4)/6.0

print(f"Fourth Order RK: y1_est = %.5f, y1_exact = %.5f, error = %.3f" % (y1,y(t0+tau_b), np.abs(y1-y(t0+tau_b))*100/y(t0+tau_b)))

time = np.linspace(t0,t0+1.6,100)
y_exact = y(time)

yk1 = y0 + time*k1
yk2 = y0 + time*k2
yk3 = y0 + time*k3
yk4 = y0 + time*k4

yk1pt = y0 + tau_b*k1
yk2pt = y0 + tau_b*k2
yk3pt = y0 + tau_b*k3
yk4pt = y0 + tau_b*k4
tau_array = np.array([t0+tau_b,t0+tau_b,t0+tau_b,t0+tau_b])
ypt_array = np.array([yk1pt,yk2pt,yk3pt,yk4pt])

tau_exact_pt = np.array([t0+tau_b])
y_exact_pt = np.array([y(t0+tau_b)])

tau_est_pt = np.array([t0+tau_b])
y_est_pt = np.array([y1])


plt.plot(time,y_exact,'k-',label='Exact')
plt.plot(time,yk1,'r--',label='y0 + t*k1')
plt.plot(time,yk2,'g--',label='y0 + t*k2')
plt.plot(time,yk3,'b--',label='y0 + t*k3')
plt.plot(time,yk4,'m--',label='y0 + t*k4')
plt.plot(tau_array,ypt_array,'ko',label='y0 + (t0+tau)*k Estimation')
plt.plot(tau_exact_pt,y_exact_pt,'ro',label='y1.4 exact')
plt.plot(tau_est_pt,y_est_pt,'bo',label='y1.4 est')
# plt.ylim(50)

plt.legend()
plt.xlabel('t')
plt.ylabel('y')
```
